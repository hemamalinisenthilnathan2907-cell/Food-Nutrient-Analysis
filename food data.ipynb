```python
data = data.drop('ID',axis=1)
```

```python
data
```

```python
top_calories = data[['Description', 'Calories']] \
    .sort_values(by='Calories', ascending=False) \
    .head(10)

plt.figure(figsize=(12,6))

plt.bar(top_calories['Description'],
        top_calories['Calories'],
        color='orange')

plt.title("Top 10 High Calorie Foods")
plt.xlabel("Food Items")
plt.ylabel("Calories")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(data['Protein'].dropna(),
         bins=20,
         color='skyblue',
         edgecolor='black')

plt.title("Protein Distribution")
plt.xlabel("Protein")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.scatter(data['TotalFat'], data['SaturatedFat'],s=10,alpha=0.5)
plt.xlabel('Total Fat')
plt.ylabel('Saturated Fat')
plt.title('Scatter Plot of Total Fat vs Saturated Fat')
```

```python
plt.figure(figsize=(10, 10))
data.hist()
```

```python
x = data['Calcium'].dropna()
y = data['Sodium'].dropna()
plt.figure(figsize=(10,6))
plt.violinplot([x, y],
               showmeans=True,
               showmedians=True)
plt.xticks([1, 2], ['Calcium', 'Sodium'])
plt.title("Violin Plot of Calcium and Sodium")
plt.ylabel("Values")
plt.show()
```

```python
df1 = data.drop('Description',axis=1)
df1
```

```python
avg_values = df1.mean()
avg_values
```

```python
plt.figure(figsize=(8, 5))

plt.bar(avg_values.index,
        avg_values.values,
        color='#c2026d')

plt.xticks(rotation=45)

plt.xlabel('Nutrients')
plt.ylabel('Average Value')
plt.title('Average Nutritional Values')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.hist(data['Calories'].dropna(),
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Distribution of Calories')

plt.show()
```

```python
corr_matrix=df1.corr()
```

```python
corr_matrix = df1.corr()
plt.figure(figsize=(15,15))
plt.imshow(corr_matrix,
           cmap='coolwarm',
           interpolation='nearest')
plt.colorbar()
plt.xticks(np.arange(len(corr_matrix.columns)),
           corr_matrix.columns,
           rotation=90)
plt.yticks(np.arange(len(corr_matrix.columns)),
           corr_matrix.columns)
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i,
                 round(corr_matrix.iloc[i, j], 2),
                 ha='center',
                 va='center',
                 color='black')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
box=plt.boxplot([
    data['Protein'].dropna(),
    data['TotalFat'].dropna(),
    data['Carbohydrate'].dropna()
],
labels=['Protein', 'TotalFat', 'Carbohydrate'],patch_artist=True)
colors = ['skyblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.ylabel('Nutrient Value')
plt.title('Distribution of Nutrients')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.stem(data['Protein'].head(20),
         linefmt='green',
         markerfmt='go',
         basefmt='black')

plt.xlabel('Food Index')
plt.ylabel('Protein')
plt.title('Stem Plot of Protein Values')

plt.show()
```

```python
x = range(10)

protein = data['Protein'].head(10)
fat = data['TotalFat'].head(10)
carbs = data['Carbohydrate'].head(10)

plt.figure(figsize=(10,6))

plt.stackplot(x,
              protein,
              fat,
              carbs,
              labels=['Protein', 'Fat', 'Carbs'],
              colors=['skyblue', 'lightgreen', 'salmon'])

plt.legend(loc='upper left')

plt.title('Nutrient Stack Plot')
plt.xlabel('Food Index')
plt.ylabel('Values')

plt.show()
```
